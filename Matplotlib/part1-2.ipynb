```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x=np.linspace(0,5,11)
```

```python
x
```

```python
y=x**2
```

```python
#Functional Methods of Plotting
plt.plot(x,y)
```

```python
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Functional Plot')
```

```python
#Subplots
plt.subplot(2,3,1)
plt.title('first')
plt.plot(x, y, 'r')
plt.subplot(2,3,2)
plt.plot(x,y,'b')
plt.subplot(2,3,3)
plt.plot(x,y,'b')
```

```python
#Object Oriented Method of Plotting
fig= plt.figure()
axes1= fig.add_axes([0,0, 1,1])
axes2= fig.add_axes([0.1, 0.6, 0.2,0.3])
```

```python
axes1.plot(x,y, 'r')
axes2.plot(x,y, 'b')
axes1.set_xlabel('X')
axes2.set_xlabel('X')
axes1.set_ylabel('Y')
axes2.set_ylabel('Y')
axes1.set_title('Larger Axis')
axes2.set_title('Smaller Axis')
fig
```

```python
#Fig Size and DPI
fig= plt.figure(figsize=(4,4), dpi=100)
axes= fig.add_axes([0,0.2, 0.4, 0.4])
axes.plot(x,y, 'g')
```

```python
fig, axes= plt.subplots(1,2, figsize=(12,1))
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
fig.savefig('diag1.png', dpi=10)
```

```python
#Legend

fig= plt.figure()
axes=fig.add_axes([0,0.1, 0.8, 0.8])
axes.plot(x,x**2, label='X Sqaure')
axes.plot(x,x**3, label='x Cube')
axes.legend(loc=0)  #best
```
