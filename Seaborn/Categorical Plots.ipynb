# Categorical Plots

```python
import seaborn as sn
```

```python
%matplotlib inline
```

```python
data= sn.load_dataset('tips')
```

```python
data.head()
```

Barplot

```python
sn.barplot(x='sex', y='total_bill', data=data)
```

```python
#using estimator param which accepts aggregate functions
import numpy as np
sn.barplot(x='sex', y='total_bill', data=data, estimator= np.std)
```

Countplot

```python
sn.countplot(x='sex', data=data)
```

Boxplot

```python
sn.boxplot(x='day', y='total_bill', data=data)
```

```python
#passing hue param with value of smoker to determine the total bill value of smokers and non-smokers on different day
sn.boxplot(x='day', y='total_bill', data=data, hue='smoker')
```

Violinplot

```python
sn.violinplot(x='day', y='total_bill', data=data)
```

```python
#with hue param
sn.violinplot(x='day', y='total_bill', data=data, hue='smoker')
```

Stripplot

```python
sn.stripplot(x='day', y='total_bill', data=data)
```

```python
#for more meaningful and clear info, using params like hue, jitter, split
sn.stripplot(x='day', y='total_bill', data=data, jitter= True)
```

```python
sn.stripplot(x='day', y='total_bill', data=data, jitter= True, hue='sex')
```

```python
sn.stripplot(x='day', y='total_bill', data=data, jitter= True, hue='sex', dodge=True)
```

Swarmplot

```python
sn.swarmplot(x='day', y='total_bill', data=data)
```

Factorplot

```python
#violinplot
sn.factorplot(x='day', y='total_bill', data=data,kind='violin')
```

```python
#barplot
sn.factorplot(x='day', y='total_bill', data=data,kind='bar')
```

```python
#box
sn.factorplot(x='day', y='total_bill', data=data,kind='box')
```
